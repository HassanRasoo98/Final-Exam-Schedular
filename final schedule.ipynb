### Select a list of subject and get personalized timetable

```python
!pip install Levenshtein
```

```python
import Levenshtein

# this function takes two inputs: a subject name user has selected from the list, and the list of all the subjects
# the function will return the subject name from the list that user selection matches closest to

# the app.py file contains a logic which reduces the time and computational complexity of this function
# suggested by gpt
def get_most_similar_target(source, target_list):
    similarities = [Levenshtein.ratio(source, target) for target in target_list]
    max_similarity_index = similarities.index(max(similarities))
    most_similar_target = target_list[max_similarity_index]
    return most_similar_target
```

```python
# example user selected subjects from the list
selected_subjects = '''AI4002 Computer Vision BS(AI) J,K 2020
MG1002 Marketing Management BS(CS) (A,B,C) BS(AI) (AB) BS(SE) (ABC) 2022
AL4002 Computer Vision Lab BS(AI) J,K 2020
AI3001 Knowledge Rep. and Reasoning BS(AI) (J,K) 2021
CS4085 MLOps BS(AI)-A, BS(CS) A 2020
MG1001 Fundamentals of Management BS(CS)-A BS(AI)-AB BS(SE)-ABC 2022'''.split('\n')

inverse_map = []

for subject in selected_subjects:
    most_similar_target = get_most_similar_target(subject, subjects)
    inverse_map.append(most_similar_target)
          
# map user selected subjects to subject names in the original list for exact match ahead  
inverse_map
```

```python
# filter and make a smaller df based on the subjects that user selected
filtered_df = df[df.apply(lambda row: any(val in row.values for val in inverse_map), axis=1)]
```

```python
df = filtered_df.drop_duplicates()
df.reset_index(drop=True)

# this df contains the rows that have the user selected subjects but we only need data of that particular subject
# so we will next remove the extra information
```

The target dataframe will have these columns: subject, day, date, time

```python
# Function to get column names for each row meeting the condition
# condition: return only that column which contains user selected subject
def get_matching_columns(row):
    return list(df.columns[row.isin(inverse_map)])[0]

# Apply the function to each row
df['matching_columns'] = df.apply(get_matching_columns, axis=1)
df.reset_index(drop=True, inplace=True)
df
# this df contains a new column named matching_columns which stores the information of which column of this particular
# row contains the information i need. we will use this to get rid of all the extra information
```

```python
# The final dataframe will have these columns: subject, day, date, time
final_df = pd.DataFrame(columns=['Subject', 'Days & Date', 'Time'])
final_df
```

```python
for i,j in df.iterrows():
    time = df['matching_columns'].iloc[i]
    final_df.at[i, 'Subject'] = df[str(time)].iloc[i]
    final_df.at[i, 'Time'] = time
    final_df.at[i, 'Days & Date'] = df['Days & Date'].iloc[i]

final_df
```

```python
# Sorting the DataFrame based on the 'Date' column
df = final_df.sort_values(by='Days & Date')
df['Date'] = pd.to_datetime(df['Days & Date'])
df['Date'].info()
df
```

```python
def std_date(date):
    # bring date to a standard format
    return date.strftime('%d-%m-%Y')

def day_name(dt):
    # return day name given a date
    return dt.day_name()

df['Days & Date'] = df['Days & Date'].apply(std_date)
# Creating a separate column for the day name
df['Day_Name'] = df['Date'].apply(day_name)
# Rearranging the columns
df = df[['Subject', 'Day_Name', 'Date', 'Time']]

# change the timings of Friday exam
condition = (df['Time'] == '1:00 to 4:00 PM') & (df['Day_Name'] == 'Friday')
column_to_modify = 'Time'
new_value = '1:30 to 4:30 PM'

df.loc[condition, column_to_modify] = new_value
df.to_json()
```
